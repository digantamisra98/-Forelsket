```python
### Import Necessary Modules
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 20]
import numpy as np
```

```python
# Variable and Function Definition
x = np.linspace(-2.5,2.5,100)
xs = np.linspace(-5.0,6.0,100)
r = np.linspace(1.0,5.0,100)
rs= np.linspace(1.0,30.0,100)
y = x * np.tanh(np.log(1+np.exp(x)))
z = 1/ (rs * np.tanh(np.log(1+np.exp(rs))))
t = np.log(r)
w = np.sin(x) + np.cos(x)
```

```python
# Plot
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, polar = True)
#ax.plot(y, 'darkblue', linewidth=200.0)
#ax.plot(y, 'maroon', linewidth = 10.0)
ax.plot(t, 'pink', linewidth=50.0)
ax.plot(z, 'k')
ax.plot(w, 'k')
ax.plot(t, 'r.', linewidth=50.0)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
#plt.savefig('x5.png',bbox_inches='tight')
plt.show()
```
